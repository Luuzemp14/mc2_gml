### Aufgabe 3 - Modell-Evaluation (4 Punkte)

Diskutiere für die vorliegende Problemstellung, welche Metriken zur Evaluation in Frage kommen.  

Welche Vor- und Nachteile haben sie?  

Was für Ansätze gibt es für die Aggregation der Metriken bei mehreren Klassen?  

Precision (Präzision):

Precision zeigt  den Anteil der korrekten positiven Vorhersagen im Verhältnis zu der Gesamtzahl der positiven Vorhersagen (sowohl korrekt als auch falsch). Die Formel für Precision lautet:

$$
\text{Precision} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Positives (FP)}}
$$

 
Hierbei sind:

True Positives (TP): Anzahl der korrekt als positiv klassifizierten Fälle.
False Positives (FP): Anzahl der fälschlicherweise als positiv klassifizierten Fälle.

Vorteile: Sehr nützlich, wenn die Kosten von falsch positiven Ergebnissen hoch sind.
Hilft zu bestimmen, wie verlässlich die positiven Vorhersagen eines Modells sind.

Nachteile:
Kann irreführend sein, wenn die Anzahl der tatsächlichen positiven Fälle sehr gering ist (unausgewogene Datensätze).
Beachtet nicht die Fälle, die fälschlicherweise als negativ klassifiziert wurden.


Recall (Sensitivität)
Recall ist ein Maß dafür, wie gut ein Klassifikationsmodell alle tatsächlich positiven Fälle erfasst. Es zeigt den Anteil der korrekten positiven Vorhersagen im Verhältnis zur Gesamtzahl der tatsächlichen positiven Fälle. Die Formel für Recall lautet:

$$
\text{Precision} = \frac{\text{True Positives (TP)}}{\text{True Positives (TP)} + \text{False Negatives (FP)}}
$$

Hierbei sind:

True Positives (TP): Anzahl der korrekt als positiv klassifizierten Fälle.
False Negatives (FN): Anzahl der fälschlicherweise als negativ klassifizierten Fälle.

Vorteile:
Wichtig, wenn die Kosten von falsch negativen Ergebnissen hoch sind (z.B. in der medizinischen Diagnostik).
Misst, wie gut das Modell alle tatsächlichen positiven Fälle identifizieren kann.

Nachteile:
Kann irreführend sein, wenn es viele falsch positive Ergebnisse gibt.
Ein hohes Recall kann durch ein Modell erreicht werden, das zu viele positive Vorhersagen macht.


F1 score: 

Der F1 Score ist eine Metrik, das insbesondere bei unausgewogenen Klassenverteilungen (wie ich sie in den test und traindaten habe) nützlich ist.
Der F1 Score erreicht seinen besten Wert bei 1 (perfekte Precision und Recall) und seinen schlechtesten bei 0. Er ist besonders hilfreich, wenn das Suchen nach einem Gleichgewicht zwischen Precision und Recall wichtig ist, insbesondere in Situationen, wo eine unausgewogene Klasse vorliegt oder wenn sich die Kosten von False Positives und False Negatives stark unterscheiden.


$$
F1 Score = 2 \times \frac{\text{Precision} \times \text{Recall}}{\text{Precision} + \text{Recall}}
$$

Vorteile:
Bietet ein ausgewogeneres Maß, das sowohl Precision als auch Recall berücksichtigt.
Besonders nützlich bei unausgewogenen Datensätzen.
Hilfreich, wenn ein Kompromiss zwischen Precision und Recall erforderlich ist.

Nachteile:
Kann schwer zu interpretieren sein, da es eine Kombination aus zwei Metriken ist.
Nicht hilfreich, wenn man sich spezifisch für nur eine der beiden Metriken (Precision oder Recall) interessiert.


Accuracy: 
Accuracy ist eine der grundlegendsten Metriken für die Beurteilung von Klassifikationen. Sie misst dabei die Anzahl aller korrekt vorhergesagten werte, also True Positiv und True Negativ im Verhältniss zu der Gesammtanzahl an Vorhersagen. 

$$ 
Accuracy = \times \frac{\text{True Positives} + \text{True Negatives}}{\text{True Positives} + \text{True Negatives} + \text{False Positives} + \text{False Negatives}}
$$

Vorteile: Einfach und intuitiv: Sie gibt einen schnellen Überblick über die Leistung eines Modells.
Effektiv für ausgewogene Datensätze: In Situationen, in denen die Klassen annähernd gleich verteilt sind, kann die Accuracy eine sinnvolle Leistungsbeurteilung liefern.

Nachteile: 
rreführend bei unausgewogenen Datensätzen: Bei stark unausgewogenen Klassenverteilungen kann ein Modell, das stets die Mehrheitsklasse vorhersagt, eine hohe Accuracy aufweisen, obwohl es möglicherweise nicht effektiv ist.
Ignoriert die Kosten unterschiedlicher Fehlklassifikationen: Die Accuracy berücksichtigt nicht, dass manche Arten von Fehlern schwerwiegender sein können als andere.

### Aufgabe 4 - 'Dummy-Klassifikatoren' (6 Punkte)

Wir wollen als Ausgangslage zwei heuristische Klassifikations-Strategien evaluieren:

1. Einen Dummy-Klassifikator, welcher immer das häufigste Label der Trainingsdaten vorhersagt.
2. Einen Dummy-Klassifikator, welcher ein zufälliges Label vorhersagt, jeweils mit der Wahrscheinlichkeit seines Vorkommens in den Trainingsdaten.

Berechne die Overall Accuracy und den ungewichteten mittleren F1-Score auf den Testdaten für beide Strategien.  

Diskutiere das Ergebnis und die resultierenden Einsichten für die folgenden Modell-basierte Ansätze.

```python
# Overall Accuracy berechnen
def calculate_accuracy(y_true, y_pred):
    """Berechnet die Overall Accuracy."""
    return np.mean(y_true == y_pred)

# funktion für f1 Score berechnen
def calculate_f1_score(y_true, y_pred, classes):
    """Berechnet den ungewichteten mittleren F1-Score."""
    f1_scores = []
    for cls in classes:
        # True Positives, False Positives, False Negatives
        tp = np.sum((y_true == cls) & (y_pred == cls))
        fp = np.sum((y_true != cls) & (y_pred == cls))
        fn = np.sum((y_true == cls) & (y_pred != cls))

        # precision und recall berechnen
        precision = tp / (tp + fp) if tp + fp > 0 else 0
        recall = tp / (tp + fn) if tp + fn > 0 else 0

        # F1-Score berechnen, wenn möglich
        f1 = 2 * ((precision * recall) / (precision + recall)) if precision + recall > 0 else 0
        f1_scores.append(f1)

    return np.mean(f1_scores)

# Liste aller Klassen
all_classes = np.unique(train_data['Class'])
```

```python
# Liste aller Klassen
all_classes = np.unique(train_data['Class'])
# Methode 1: Häufigstes Label

# Bestimmen des häufigsten Labels im Trainingsdatensatz
most_common_label = train_data['Class'].mode()[0]

# Erzeugen der Vorhersagen mit dem häufigsten Label für den Testdatensatz
predictions_most_common = np.full(len(test_data), most_common_label)

# Berechnung der Metriken für die erste Strategie mit manuellen Methoden
accuracy_most_common = calculate_accuracy(test_data['Class'], predictions_most_common)
f1_score_most_common = calculate_f1_score(test_data['Class'], predictions_most_common, all_classes)

accuracy_most_common, f1_score_most_common
```

```python
# Methode 2: Zufälliges Label

# Relative Häufigkeiten der Labels in den Trainingsdaten
class_probabilities = train_data['Class'].value_counts(normalize=True)

np.random.seed(42)  # Für reproduzierbare Ergebnisse
predictions_random = np.random.choice(class_probabilities.index, 
                                      size=len(test_data), 
                                      p=class_probabilities.values)

# Berechnung der Metriken für die zweite Strategie
accuracy_random = calculate_accuracy(test_data['Class'], predictions_random)
f1_score_random = calculate_f1_score(test_data['Class'], predictions_random, all_classes)

accuracy_random, f1_score_random
```

Accuracy: Diese Strategie könnte eine anständige Genauigkeit erzielen, wenn die Daten stark unausgewogen sind und die Mehrheit der Fälle das häufigste Label aufweist. ### Aufgabe 6 - Multi-Layer Perceptron (15 Punkte)

Implementiere ein Multi-Layer Perceptron mit Sigmoid Aktivierungsfunktionen zur Multi-Klass Klassifikation.

Dabei sollen min. Architektur (# layers / units per layer), Regularisierung ($\alpha$), learning rate ($\eta$), und Batch-Size konfigurierbar sein. 

Initialisiere die Gewichte zufällig mit kleinen Werten, z.B. aus dem Intervall $[-0.7, 0.7]$. Dabei sollst du sicherstellen, dass der Fit der Modelle reproduzierbar ist. Setze also _random seeds_ um die Initialisierung der Gewichte reproduzierbar zu machen [siehe numpy](https://numpy.org/doc/stable/reference/random/index.html#random-quick-start).  

Ergänze dazu die vorgegebene Klasse.

#### Kostenfunktion

Es soll folgende Kostenfunktion $J$ minimiert werden (_cross-entropy_ mit Regularisierung):

\begin{align}
J(\mathbf{w}) &= \frac{1}{N}\sum_{i=1}^{N} \sum_{k=1}^{K} \Big(- y_k^{(i)} \log \hat{y}_k^{(i)} \Big) + C \\
C &= \frac{\alpha}{2} \sum_{l=1}^{L}\sum_{j=1}^{u^{(l)}} \sum_{i=1}^{u^{(l-1)}} (w^{(l)}_{ij})^2
\end{align}

#### Notation

- $N$: Anzahl Datenpunkte
- $L$: Anzahl Layer (ohne Input-Layer)
- $K$: Anzahl Klassen
- $p$: Anzahl Input Features
- $\alpha$: Regularisierungsfaktor
- $\mathbf{a}^{(l)}$ Aktivierungen im Layer $l$ (l=0 ist Input-Layer)
- $\mathbf{z}^{(l)}$ pre-activations / logits im Layer $l$

- $\mathbf{y}^{(i)} \in \{0, 1\}$: One-hot encoded Vektor mit Dimensionalität $\mathbb{R}^K$, Eintrag 1 indiziert die Klassenzugehörigkeit
- $\hat{y}_k^{(i)}$ Prediction für Klasse $k$, $P(Y=k| x^{(i)})$ 
- $w^{(l)}_{ij}$: Gewicht / Parameter $ij$ von Layer $l$
- $W^{(l)}$ Parameter von Layer $l$ mit $\in \mathbb{R}^{u^{(l-1)} \times u^{(l)}}$ oder mit Bias $\mathbb{R}^{(u^{(l-1)}+1) \times u^{(l)}}$
- $u^{(l)}$: Anzahl Nodes (units) im Layer $l$, ohne Bias Node
- $g()$: Aktivierungsfunktion


#### Forward Pass

Exemplarisch für einen Datenpunk $\mathbf{x}^{(i)}$. Wir nehmen an, dass der Bias als 1-er Spalte eingefügt wurde $x_0^{(i)}=1$.

\begin{align*}
\mathbf{a}^{(0)} &= \mathbf{x}^{(i)} \\
\mathbf{z}^{(1)} &= \mathbf{a}^{(0)T}W^{(1)} \\
\mathbf{a}^{(1)} &= g(\mathbf{z}^{(1)}) \\
.. &= .. \\
\mathbf{z}^{(l)} &= \mathbf{a}^{(l-1)T}W^{(l)} \\
\mathbf{a}^{(l)} &= g(\mathbf{z}^{(l)}) \\
.. &= .. \\
\mathbf{\hat{y}}_k^{(i)} &= a_k^{(L)} \\
g(\mathbf{z}^{(l)}) &=  \frac{1}{1+e^{-z^{(l)}}} & \text{Sigmoid Aktivierungsfunktion für Hidden-Layers } (l \lt L)  \\
a_i^{(L)} &= \sigma(\mathbf{z}^{(L)})_i = \frac{e^{z_i}}{\sum_{j=1}^K e^{z_j}} & \text{ Softmax-Funktion for } l = L
\end{align*}

#### Backpropagation

Backpropagation ist der zentrale Algorithmus beim Trainieren von Neuronalen Netzwerken um die partiellen Ableitung der Loss-Funktion bezüglich der Parameter zu berechnen. Dabei wird der Gradient der Loss-Funktion vom Output des Netzwerkes mit Hilfe der Kettenregel zurückpropagiert. Du sollst diesen Algorithmus anwenden, um alle $\frac{\partial J}{\partial \mathbf{w_{ij}^{(l)}}}$ zu bestimmen. Folgende Zusammenhänge sind hilfreich bei der Implementation:


\begin{align*}
\frac{\partial a^{(l)}}{\partial z^{(l)}} &= g(z^{(l)})(1 - g(z^{(l)})) = a^{(l)} \odot (1- a^{(l)}) & \text{for } l \lt L\\
\frac{\partial z^{(l)}}{\partial a^{(l-1)}} &= W^{(l)} \\
\frac{\partial z^{(l)}}{\partial W^{(l)}} &= a_i^{(l-1)}
\end{align*}


Man kann den Gradienten der Kostenfunktion bezüglich der Gewichte $W$ rekursiv bestimmen:

\begin{align*}
\frac{\partial J}{\partial \mathbf{z}^{(L)}} &= \delta^{(L)} = \mathbf{\hat{y}} - \mathbf{y} & \text{inizialisiere backprop hiermit. }\\
\delta^{(l)} &= \Big(\frac{\partial z^{(l+1)}}{\partial a^{(l)}} \delta^{(l+1)} \Big) \odot \frac{\partial a^{(l)}}{\partial z^{(l)}} & \text{berechne die backprop-Schritte iterativ für } l \lt L \\
\frac{\partial J}{\partial W^{(l)}} &=  \Big(\frac{\partial z^{(l)}}{\partial W^{(l)}}\Big)^T \delta^{(l)}  & \text{dies sind die Gradienten der Gewichte!}
\end{align*}

#### Vektorisierung

Wir können den Forward-Pass und die Backpropagation effizient berechnen, indem wir die Operationen vektorisieren. Wir berechnen die Aktivierungen direkt für alle Datenpunkte $N$. Die Bias-Vektoren $\mathbf{b}$ werden direkt in die Aktivierungs-Matrizen hinzugefügt. Auch die Backpropagation lässt sich vektorisieren, wir berechnen dann den durchschnittlichen Gradienten über alle Datenpunkte. Ohne Vektorisierung wird deine Implementation zu langsam sein. Hier der Ablauf für die Forward-Propagation:


\begin{align*}
\mathbf{X} & \in \mathbb{R}^{N \times (p + 1)} \\
\mathbf{A}^{(l)} & \in \mathbb{R}^{N \times (u^{(l)} + 1)} \\
\mathbf{Z}^{(l)} & \in \mathbb{R}^{N \times (u^{(l)})} \\
\mathbf{W}^{(l)} & \in \mathbb{R}^{(u^{(l-1)} + 1) \times u^{(l)}} \\
\mathbf{b}^{(l)} & \in \mathbb{R}^{u^{(l)} \times 1} \\
\mathbf{Z}^{(1)} &= \mathbf{X} \mathbf{W}^{(1)} \\
\mathbf{A}^{(1)} &= g( \mathbf{Z}^{(1)}) & \text{add bias node if necessary}\\
... \hspace{4pt} &= \hspace{4pt} ... \\
\mathbf{A}^{(L)} &= g( \mathbf{Z}^{(L)}) & \text{no bias at output}\\
\end{align*}

```python
import copy
from itertools import islice
from typing import List, Tuple, Union, Iterable

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from tqdm.notebook import tqdm

from sklearn.base import BaseEstimator
from sklearn import metrics
from sklearn.preprocessing import LabelBinarizer, OneHotEncoder


class MLPClassifier(BaseEstimator):
    """ Multi-Layer Perceptron for Classifikation
    
        Args:
            num_nodes_per_layer: List with the number of nodes per layer
                 - including input-layer (features)
                 - no bias nodes should be included
                 - example: [10, 5, 2] -> 10 Features, 5 Nodes in 1st Hidden Layer, 2 Output Nodes
    
            lr: initial learning rate (float)
    
            alpha: L2 Regularization strength (float)
            
            batch_size: number of samples to use for one update 
                - set to None for batch gradient descent (all samples)
        
            max_num_steps: max number of steps (gradient updates)
            
            weights_init_range: Tuple of two floats. 
                - The interval on which the weights will be randomly initialized
                
            adaptive_lr_constant: Adpative learning rate constant
                - lr_new = lr * exp(-adaptive_lr_constant * step)
                - disable with 0.0
    
            min_lr: Minimum Learning Rate, relevant when using adaptive_lr_constant > 0.0
            
            random_seed: random seed for weights initialization (int)
    """
    def __init__(
        self, 
        num_nodes_per_layer: List[int],
        lr: float=0.01,
        alpha: float=0.0,
        batch_size: Union[int, None]=None,
        max_num_steps: int=1000,
        weights_init_range: Tuple[float, float]=(-.7, .7),
        adaptive_lr_constant: float=0.0,
        min_lr: float=1e-4,
        random_seed: int=123
    ):

        ### BEGIN SOLUTION
        self.num_nodes_per_layer = num_nodes_per_layer
        self.lr = lr
        self.alpha = alpha
        self.batch_size = batch_size
        self.max_num_steps = max_num_steps
        self.weights_init_range = weights_init_range
        self.adaptive_lr_constant = adaptive_lr_constant
        self.min_lr = min_lr
        self.random_seed = random_seed
        self.weights_ = []
        self._initialize_weights()
        self.cost_history = []
        ### END SOLUTION
        
    def _initialize_weights(self):
        """ Initialize Model-Parameters W """
        ### BEGIN SOLUTION
        # Seed setzen 
        np.random.seed(self.random_seed)        
        
        # For loop, um die Gewichte für alle Layer zu initialisiert
        for i in range(len(self.num_nodes_per_layer) - 1):
            # Form der Gewichtsmatrix (+1 wegen Bias)
            layer_shapes = self.num_nodes_per_layer[i] + 1, self.num_nodes_per_layer[i+1]
            self.weights_.append(np.random.uniform(self.weights_init_range[0], self.weights_init_range[1],layer_shapes))

        ### END SOLUTION

    
    def forward(self, X: np.ndarray) -> List[np.ndarray]:
        """ forward pass - calculate all activations
            Args:
                X: Input-Activations [num_samples, num_features]
    
            Returns: List of Layer-Activations
        """
        ### BEGIN SOLUTION
        # activations ist eine Liste, die die Activations für alle Layer enthält
        activations = [X]
        for i, weights in enumerate(self.weights_):
            # Eine Spalte mit 1 an die Input-Activations anbringen für Bias 
            bias_ones = np.hstack((np.ones((activations[-1].shape[0], 1)), activations[-1]))
            output_z = bias_ones @ weights
            # Sigmoid als activation verwenden für alle Layer außer dem letzten
            if i < len(self.weights_)-1:
                type_act = self.sigmoid(output_z)
            # für den outputlayer Softmax als activatoin verwenden
            else:
                type_act = self.softmax(output_z)
            activations.append(type_act)
        return activations
        ### END SOLUTION
    
    def backward(self, activations: List[np.ndarray], delta_upstream: np.ndarray) -> List[np.ndarray]:
        """ backward pass / backpropagation - Calculate partial derivatives of model weights
            Args:
                activations: List of Layer Activations, Output of forward()
        
                delta_upstream: Matrix with partial derivatives of Loss functions wrt. 
                    logits of last layer[num_samples, num_outputs]
    
            Returns: List of Layer-Gradients
        """
        ### BEGIN SOLUTION
        #fehlerterm delta
        delta = delta_upstream

        # Liste für gradients, welche die gleiche länge wie die weights hat 
        gradients = [None] * len(self.weights_)
        
        # For loop, um die gradients für alle Layer zu berechnen
        for i in range(len(self.weights_) - 1, -1, -1):
            # Spalte mit 1 an die Activations anhängen für Bias
            bias_ones = np.hstack((np.ones((activations[i].shape[0], 1)), activations[i]))
            gradients[i] = ((bias_ones.T @ delta) / delta.shape[0]) + self.alpha * np.hstack((np.zeros((self.weights_[i].shape[0], 1)), self.weights_[i][:, 1:]))
            if i > 0:
                delta = (delta @ self.weights_[i][1:].T) * (activations[i] * (1 - activations[i]))
        return gradients
    
        ### END SOLUTION
    
    def fit(
            self,
            X: np.ndarray, y: np.ndarray,
            X_val: Union[np.ndarray, None]=None,
            y_val: Union[np.ndarray, None]=None,
            validate_after_every_num_steps: int=20) -> None:
        """ Fit Model Parameters
            Args:
                X, y: Training-Data
                X_val, y_val: (Optional) Validation-Data
                validate_after_every_num_steps: (Optional) Validation after every xth step
        """
        ### BEGIN SOLUTION

        # One-hot encoding für y
        label_binarizer = LabelBinarizer()
        y_one_hot = label_binarizer.fit_transform(y)
        if X_val is not None and y_val is not None:
            y_val_one_hot = label_binarizer.transform(y_val)
        # Batches von trainingsdaten generieren
        batch_generator = self.generate_batches(X, y_one_hot, self.batch_size)
        for step in tqdm(range(self.max_num_steps)):
            X_batch, y_batch = next(batch_generator)
            activations = self.forward(X_batch)
            y_predicted = activations[-1]
            # cost berechnen
            cost = self.calculate_cost(y_batch, y_predicted) 
            #cost history erstellen
            delta_upstream = self.calculate_gradient_of_cost_function(y_batch, y_predicted)
            gradients = self.backward(activations, delta_upstream)
            # uptate weights
            for i in range(len(self.weights_)):
                self.weights_[i] -= self.lr * gradients[i]
            # learning rate
            if self.adaptive_lr_constant > 0.0:
                self.lr = max(self.min_lr, self.lr * np.exp(-self.adaptive_lr_constant * step))
            self.cost_history.append(cost)
        ### END SOLUTION
        
    def calculate_cost(self, y_true: np.ndarray, y_pred: np.ndarray) -> float:
        """ Calculate Cross-Entropy cost """
        ### BEGIN SOLUTION
        return -np.sum(y_true * np.log(y_pred + 1e-7)) / y_true.shape[0]
        ### END SOLUTION
    
    def calculate_gradient_of_cost_function(self, y_true: np.ndarray, y_pred: np.ndarray) -> np.ndarray:
        """ Calculate gradient of cost function w.r.t logits dJ/dZ """
        ### BEGIN SOLUTION
        return y_pred - y_true

        ### END SOLUTION
    
    def softmax(self, x: np.ndarray) -> np.ndarray:
        """ Softmax along last dimension: x is of shape [num_samples, num_classes] """
        ### BEGIN SOLUTION
        x_expo = np.exp(x - np.max(x, axis=-1, keepdims=True))
        return x_expo / x_expo.sum(axis=-1, keepdims=True)
        ### END SOLUTION

    def sigmoid(self, x: np.ndarray) -> np.ndarray:
        """ Sigmoid activation function """
        ### BEGIN SOLUTION
        return 1 / (1 + np.exp(-x))
        ### END SOLUTION
    
    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        """ All class probabilities per sample
            -> input shape  [num_samples, num_features]
            -> output shape [num_samples, num_classes]
        """
        ### BEGIN SOLUTION
        return self.forward(X)[-1]
        ### END SOLUTION        

    def predict(self, X: np.ndarray) -> np.ndarray:
        """ Return most likely class per sample
           -> input shape  [num_samples, num_features]
           -> output shape [num_samples, 1]
        """
        ### BEGIN SOLUTION
        return np.argmax(self.predict_proba(X), axis=-1)
        ### END SOLUTION

    def score(self, X: np.ndarray, y: np.ndarray) -> np.ndarray:
        """ Calculate Accuracy """
        ### BEGIN SOLUTION
        return np.mean(np.squeeze(y) == np.squeeze(self.predict(X)))
        ### END SOLUTION

    def generate_batches(
            self, X: np.ndarray, y: np.ndarray,
            batch_size: Union[int, None],
            shuffle: bool=True) -> Iterable[Tuple[np.ndarray, np.ndarray]]:
        """ Generate batches of data [BATCH_SIZE, NUM_FEATURES]"""
    
        ### BEGIN SOLUTION

        # Batch Gradient Descent
        if not batch_size:
            batch_size = X.shape[0]
            while True:
                yield X, y
        
        # Flexible Batch-Size that works with batch_size != None
        while True:
            # randomly shuffle input data
            shuffled_indices = np.arange(X.shape[0])
            if shuffle:
                np.random.shuffle(shuffled_indices)
            
            # iterate over batches
            iter_indices = np.arange(X.shape[0])
            slices = islice(iter_indices, 0, None, batch_size)

            # select batches
            for start_index in slices:
                indices_to_select = shuffled_indices[start_index: start_index+batch_size]
                yield X[indices_to_select, :], y[indices_to_select, :]
        ### END SOLUTION
        
    def grad_check(
            self, X: np.ndarray, y: np.ndarray,
            epsilon: float=0.0001,
            decimal: int=3,
            verbose: bool=False) -> None:
        """Compare the gradient with finite differences around current point in parameter space
            Args:
                X, y: Data
                epsilon: step to take around current parameters
                decimal: verify equality if approximaton is within +- 1e-<decimal>
        """
        # calculate gradient
        activations = self.forward(X)
        y_hat = activations[-1]
        delta_upstream = self.calculate_gradient_of_cost_function(y, y_hat)
        gradients = self.backward(activations, delta_upstream)
        
        # approximate gradient
        gradients_approx = copy.deepcopy(gradients)
        original_layers = copy.deepcopy(self.weights_)
        
        # Iterate over each parameter of the network
        for i, weights in enumerate(self.weights_):
            for j, _ in enumerate(weights.flat):
                
                # generate copy of original parameters for modification
                modified_layers = copy.deepcopy(original_layers)
          
                # assign modified layers for use in other methods
                self.weights_ = modified_layers
                
                # J(theta + epsilon)
                modified_layers[i].flat[j] += epsilon
                y_hat = self.predict_proba(X)
                cost_plus_epsilon = self.calculate_cost(y, y_hat)
                
                # J(theta - epsilon)
                modified_layers[i].flat[j] -= 2 * epsilon           
                y_hat = self.predict_proba(X)
                cost_minus_epsilon = self.calculate_cost(y, y_hat)
                
                # Approx gradient with:
                # (J(theta + epsilon) - J(theta - epsilon)) / (2 * epsilon)
                grad_approx = (
                    (cost_plus_epsilon - cost_minus_epsilon) / 
                    (2 * epsilon))

                gradients_approx[i].flat[j] = grad_approx
                
        # reset layers attribute
        self.weights_ = copy.deepcopy(original_layers)
        
        # normalize gradients
        gradients_approx = [x / np.linalg.norm(x) for x in gradients_approx]
        gradients = [x / np.linalg.norm(x) for x in gradients]
                

        if verbose:
            print('approx : ', gradients_approx)
            print('calc : ', gradients )
        
        for approx, calculated in zip(gradients_approx, gradients):
            np.testing.assert_array_almost_equal(approx, calculated, decimal=decimal)
        print(f"Gradients within +- 1e-{decimal} of approximated gradient!")

    # Additional Methods for your convenience (optional)
    ### BEGIN SOLUTION
    
    ### END SOLUTION
```

```python
import numpy as np
from sklearn.base import BaseEstimator
from sklearn.preprocessing import LabelBinarizer
from scipy.optimize import minimize

class MLPClassifierExtended(BaseEstimator):
    def __init__(
        self, 
        num_nodes_per_layer: List[int],
        lr: float=0.01,
        alpha: float=0.0,
        batch_size: Union[int, None]=None,
        max_num_steps: int=1000,
        weights_init_range: Tuple[float, float]=(-.7, .7),
        adaptive_lr_constant: float=0.0,
        min_lr: float=1e-4,
        random_seed: int=123
    ):
        self.num_nodes_per_layer = num_nodes_per_layer
        self.lr = lr
        self.alpha = alpha
        self.batch_size = batch_size
        self.max_num_steps = max_num_steps
        self.weights_init_range = weights_init_range
        self.adaptive_lr_constant = adaptive_lr_constant
        self.min_lr = min_lr
        self.random_seed = random_seed
        self.weights_ = []
        self._initialize_weights()
        self.cost_history = []

    def _initialize_weights(self):
        np.random.seed(self.random_seed)        
        for i in range(len(self.num_nodes_per_layer) - 1):
            layer_shapes = self.num_nodes_per_layer[i] + 1, self.num_nodes_per_layer[i+1]
            self.weights_.append(np.random.uniform(self.weights_init_range[0], self.weights_init_range[1], layer_shapes))

    def forward(self, X: np.ndarray) -> List[np.ndarray]:
        activations = [X]
        for i, weights in enumerate(self.weights_):
            bias_ones = np.hstack((np.ones((activations[-1].shape[0], 1)), activations[-1]))
            output_z = bias_ones @ weights
            if i < len(self.weights_)-1:
                type_act = self.sigmoid(output_z)
            else:
                type_act = self.softmax(output_z)
            activations.append(type_act)
        return activations

    def backward(self, activations: List[np.ndarray], delta_upstream: np.ndarray) -> List[np.ndarray]:
        gradients = [None] * len(self.weights_)
        delta = delta_upstream
        for i in range(len(self.weights_) - 1, -1, -1):
            bias_ones = np.hstack((np.ones((activations[i].shape[0], 1)), activations[i]))
            gradients[i] = ((bias_ones.T @ delta) / delta.shape[0]) + self.alpha * np.hstack((np.zeros((self.weights_[i].shape[0], 1)), self.weights_[i][:, 1:]))
            if i > 0:
                delta = (delta @ self.weights_[i][1:].T) * (activations[i] * (1 - activations[i]))
        return gradients

    def calculate_cost(self, y_true: np.ndarray, y_pred: np.ndarray) -> float:
        return -np.sum(y_true * np.log(y_pred + 1e-7)) / y_true.shape[0]

    def calculate_gradient_of_cost_function(self, y_true: np.ndarray, y_pred: np.ndarray) -> np.ndarray:
        return y_pred - y_true

    def softmax(self, x: np.ndarray) -> np.ndarray:
        x_expo = np.exp(x - np.max(x, axis=-1, keepdims=True))
        return x_expo / x_expo.sum(axis=-1, keepdims=True)

    def sigmoid(self, x: np.ndarray) -> np.ndarray:
        x_clip = np.clip(x, -500, 500)
        return 1 / (1 + np.exp(-x_clip))

    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        return self.forward(X)[-1]

    def predict(self, X: np.ndarray) -> np.ndarray:
        return np.argmax(self.predict_proba(X), axis=-1)

    def score(self, X: np.ndarray, y: np.ndarray) -> np.ndarray:
        return np.mean(np.squeeze(y) == np.squeeze(self.predict(X)))

    def pack_parameters(self):
        packed_parameters = np.concatenate([w.ravel() for w in self.weights_])
        return packed_parameters

    def unpack_parameters(self, packed_parameters):
        shapes = [w.shape for w in self.weights_]
        num_elements = [np.prod(s) for s in shapes]
        unpacked_parameters = []
        start = 0
        for shape, num in zip(shapes, num_elements):
            end = start + num
            unpacked_parameters.append(packed_parameters[start:end].reshape(shape))
            start = end
        return unpacked_parameters

    def cost_function(self, packed_parameters, X, y):
        self.weights_ = self.unpack_parameters(packed_parameters)
        activations = self.forward(X)
        cost = self.calculate_cost(y, activations[-1])
        delta_upstream = self.calculate_gradient_of_cost_function(y, activations[-1])
        gradients = self.backward(activations, delta_upstream)
        grad_packed = np.concatenate([g.ravel() for g in gradients])
        return cost, grad_packed

    def fit(self, X, y, X_val=None, y_val=None, validate_after_every_num_steps=20):
        label_binarizer = LabelBinarizer()
        y_one_hot = label_binarizer.fit_transform(y)
        if X_val is not None and y_val is not None:
            y_val_one_hot = label_binarizer.transform(y_val)
        packed_initial_parameters = self.pack_parameters()
        result = minimize(fun=self.cost_function,
                          x0=packed_initial_parameters,
                          args=(X, y_one_hot),
                          method='L-BFGS-B',
                          jac=True,
                          options={'maxiter': self.max_num_steps})
        self.weights_ = self.unpack_parameters(result.x)
```
